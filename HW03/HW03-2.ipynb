```python
'''找出2010-2018年各行業受僱員工每人每月平均總薪資最高及最低，及其平均與折線圖'''
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

'''抓資料'''
df = pd.read_csv('https://quality.data.gov.tw/dq_download_csv.php?nid=9634&md5_url=7b41ec11a497a37184b82402be86eda5')
'''欄位資料'''
dc = np.delete(np.array(df.columns),0)

def d(x):
    '''每年最高及最低薪的產業，平均'''
    dn = np.delete(np.array(df.iloc[x]),0)
    dn = np.array(dn , dtype=np.float)
    max_job = dc[dn.argmax()]
    min_job = dc[dn.argmin()]
    mean = np.mean(dn)
    return max_job , min_job ,mean

'''找出2010到2018年最高薪與最低新的產業，及平均'''
mean = []
for i in range(390,507,13):
    max_min_mean = d(i)
    mean.append(max_min_mean[2])
    print(str(int(i/13+1980))+'年最高新產業：'+max_min_mean[0]+' ; 最低新產業：'+max_min_mean[1] + ' ; 平均：'+ str(max_min_mean[2]))
```

```python
'''10年來最低薪與最高新產業的折線圖'''
x = [i for i in range(2010,2019)]
y1 = []
y2 = []
for i in range(390,507,13):
    y1.append(float(df.電力及燃氣供應業_Electricity_and_gas_supply[i]))
    y2.append(float(df.教育業_Education[i]))

plt.figure(figsize = (10,5) )
plt.xticks(np.arange(2010,2019,1))
plt.yticks(np.arange(20000,110000,5000))
plt.plot(x,y1,label = 'Electricity and gassupply')
plt.plot(x,y2,label = 'Education')
plt.plot(x,mean,label = 'average')
plt.legend(loc = 'best');
```
