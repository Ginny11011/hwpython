```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
mpl.rc('font' , family = 'Noto Sans CJK TC')
```

匯入資料

```python
df = pd.read_csv('http://stats.moe.gov.tw/files/detail/107/107_students.csv')
```

處理資料

```python
columname = df.loc[1].values
df = df[2:]
df.columns = columname
df
```

篩出心理相關科系的學士班

```python
psy = df[(df['科系名稱'].str.contains('心理')) & (df['等級別'] == 'B 學士')]
```

```python
psy = psy.loc[:,'學校名稱':'女生計'].copy()
```

```python
psy
```

處理缺失值

```python
psy.男生計 = psy.男生計.str.replace('-','0')
psy.女生計 = psy.女生計.str.replace('-','0')
psy['男生計'] = psy['男生計'].astype('int')
psy['女生計'] = psy['女生計'].astype('int')
```

計算各校性別比

```python
psy['性別比'] = psy['男生計'] / psy['女生計']
```

```python
psy
```

繪製圖表

```python
psy1 = psy.sort_values(by = '性別比', ascending=True)
psy1.plot.bar(x = '學校名稱', y = '性別比', color ='b')
```

計算全台心理相關科系學士班性別比

```python
Sex_ratio = sum(psy.男生計)/sum(psy.女生計) *100
Sex_ratio
```
