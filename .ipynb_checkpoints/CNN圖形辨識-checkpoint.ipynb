# 1. 初始準備

```python
%env KERAS_BACKEND=tensorflow
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

# 2. 讀入 MNIST 數據庫

```python
from keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

# 3. 整理輸入輸出格式

如果你還記得, 我們每筆輸入資料都是 28x28 的陣列, CNN 其實就是吃「圖」的, 所以基本上不用像之前把每筆資料拉平。「但。是。」平常的圖都有 R, G, B 三個 channels, 每個 channel 都是一個矩陣, 也就是一張圖可能是三個矩陣! 我們是灰階, 也就是只有一個 channel。但這件事也要明確的告訴 Keras。

換句話說, 我們的輸入每筆資料型式要從 (28, 28) 換成 (28, 28, 1)!

## 3-1 輸入格式

```python
x_train[1].shape
```

```python
x_train = x_train.reshape(60000, 28, 28, 1)
x_test  = x_test.reshape(10000, 28, 28, 1)
```

```python
x_train[1].shape
```

標準化

```python
x_train = x_train/255
x_test = x_test/255
```

## 3-2 輸出格式

```python
from keras.utils import np_utils
```

變成10

```python
y_train = np_utils.to_categorical(y_train, 10)
y_test = np_utils.to_categorical(y_test, 10)
```

# 4. 打造CNN！

## 4-1 讀入相關套件

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten
from keras.layers import Conv2D, MaxPool2D
from keras.optimizers import SGD
from keras.optimizers import Adadelta
```

## 4-2 建立神經網路架構

CNN 我們一樣要決定用幾層的 CNN, 然後是不是每次都要做 max-pooling。再來就是拉平、送入標準神經網路 (再度要決定幾層、幾個神經元)。

### 4-2-1 上課版本(model_1)


**步驟**
* 建立空白神經網路
* convolution + max-pooling：共做 <span style="color:red;">3</span> 次 ，filter 大小都是 <span style="color:red;">5 x 5</span>，max-pooling 都用 <span style="color:red;">2 x 2</span> 為一小區塊
* 拉平
* 隱藏層：加上 <span style="color:red;">1</span> 個隱藏層，共使用 <span style="color:red;">9</span> 個神經元。
* 組裝：loss funtion為<span style="color:red;"> mse</span> ，optimizer為<span style="color:red;"> SGD</span>
* 訓練神經網路
* 查看測試資料結果

```python
#讀入空白神經網絡
model_1 = Sequential()
#convolution + max-pooling
model_1.add(Conv2D(4, (5, 5), padding='same', input_shape=(28,28,1)))
model_1.add(Activation('relu'))
model_1.add(MaxPool2D(pool_size=(2, 2)))

model_1.add(Conv2D(8, (5, 5), padding='same'))
model_1.add(Activation('relu'))
model_1.add(MaxPool2D(pool_size=(2, 2)))

model_1.add(Conv2D(16, (5, 5), padding='same'))
model_1.add(Activation('relu'))
model_1.add(MaxPool2D(pool_size=(2, 2)))

#拉平
model_1.add(Flatten())
#隱藏層
model_1.add(Dense(9))
model_1.add(Activation('relu'))
#輸出
model_1.add(Dense(10))
model_1.add(Activation('softmax'))
#組裝
model_1.compile(loss='mse', optimizer=SGD(lr=0.07), metrics=['accuracy'])
#檢視神經網路
model_1.summary()
#訓練
model_1_out = model_1.fit(x_train, y_train, batch_size=100, epochs=10)
#測試資料正確率
score_1 = model_1.evaluate(x_test,y_test)
print('loss :', score_1[0],
      '正確率:', score_1[1])
```

**結論**


### 4-2-2 更改loss function & optimizer(model_2)


**步驟**
* 建立空白神經網路
* convolution + max-pooling：共做 <span style="color:red;">3</span> 次 ，filter 大小都是 <span style="color:red;">5 x 5</span>，max-pooling 都用 <span style="color:red;">2 x 2</span> 為一小區塊
* 拉平
* 隱藏層：加上 <span style="color:red;">1</span> 個隱藏層，共使用 <span style="color:red;">9</span> 個神經元。
* 組裝：loss funtion為<span style="color:blue;"> categorical_crossentropy</span> ，optimizer為<span style="color:blue;"> Adadelta</span>
* 訓練神經網路
* 查看測試資料結果

```python
#讀入空白神經網絡
model_2 = Sequential()
#convolution + max-pooling
model_2.add(Conv2D(4, (5, 5), padding='same', input_shape=(28,28,1)))
model_2.add(Activation('relu'))
model_2.add(MaxPool2D(pool_size=(2, 2)))

model_2.add(Conv2D(8, (5, 5), padding='same'))
model_2.add(Activation('relu'))
model_2.add(MaxPool2D(pool_size=(2, 2)))

model_2.add(Conv2D(16, (5, 5), padding='same'))
model_2.add(Activation('relu'))
model_2.add(MaxPool2D(pool_size=(2, 2)))

#拉平
model_2.add(Flatten())
#隱藏層
model_2.add(Dense(9))
model_2.add(Activation('relu'))
#輸出
model_2.add(Dense(10))
model_2.add(Activation('softmax'))
#組裝
# model_1.compile(loss='mse', optimizer=SGD(lr=0.07), metrics=['accuracy'])
model_2.compile(loss="categorical_crossentropy", optimizer=Adadelta(), metrics=['accuracy'])
#檢視神經網路
model_2.summary()
#訓練
model_2_out = model_2.fit(x_train, y_train, batch_size=100, epochs=10)
#測試資料正確率
score_2 = model_2.evaluate(x_test,y_test)
print('loss :', score_2[0],
      '正確率:', score_2[1])
```

### 4-2-3 convolution層數(model_3)

步驟

* 建立空白神經網路
* convolution + max-pooling：共做 <span style="color:blue;">4</span> 次 ，filter 大小都是 <span style="color:red;">5 x 5</span>，max-pooling 都用 <span style="color:red;">2 x 2</span> 為一小區塊
* 拉平
* 隱藏層：加上 <span style="color:red;">1</span> 個隱藏層，共使用 <span style="color:red;">9</span> 個神經元。
* 組裝：loss funtion為<span style="color:red;"> categorical_crossentropy</span> ，optimizer為<span style="color:red;"> Adadelta</span>
* 訓練神經網路
* 查看測試資料結果

```python
#讀入空白神經網絡
model_3 = Sequential()
#convolution + max-pooling
model_3.add(Conv2D(4, (5, 5), padding='same', input_shape=(28,28,1)))
model_3.add(Activation('relu'))
model_3.add(MaxPool2D(pool_size=(2, 2)))

model_3.add(Conv2D(8, (5, 5), padding='same'))
model_3.add(Activation('relu'))
model_3.add(MaxPool2D(pool_size=(2, 2)))

model_3.add(Conv2D(16, (5, 5), padding='same'))
model_3.add(Activation('relu'))
model_3.add(MaxPool2D(pool_size=(2, 2)))

model_3.add(Conv2D(32, (5, 5), padding='same'))
model_3.add(Activation('relu'))
model_3.add(MaxPool2D(pool_size=(2, 2)))

#拉平
model_3.add(Flatten())
#隱藏層
model_3.add(Dense(9))
model_3.add(Activation('relu'))
#輸出
model_3.add(Dense(10))
model_3.add(Activation('softmax'))
#組裝
# model_1.compile(loss='mse', optimizer=SGD(lr=0.07), metrics=['accuracy'])
model_3.compile(loss="categorical_crossentropy", optimizer=Adadelta(), metrics=['accuracy'])
#檢視神經網路
model_3.summary()
#訓練
model_3_out = model_3.fit(x_train, y_train, batch_size=100, epochs=10)
#測試資料正確率
score_3 = model_3.evaluate(x_test,y_test)
print('loss :', score_3[0],
      '正確率:', score_3[1])
```

### 4-2-4 更改filter大小(model_4)


**步驟**
步驟

* 建立空白神經網路
* convolution + max-pooling：共做 <span style="color:red;">3</span> 次 ，filter 大小都是 <span style="color:blue;">4 x 4</span>，max-pooling 都用 <span style="color:red;">2 x 2</span> 為一小區塊
* 拉平
* 隱藏層：加上 <span style="color:red;">1</span> 個隱藏層，共使用 <span style="color:red;">9</span> 個神經元。
* 組裝：loss funtion為<span style="color:red;"> categorical_crossentropy</span> ，optimizer為<span style="color:red;"> Adadelta</span>
* 訓練神經網路
* 查看測試資料結果

```python
#讀入空白神經網絡
model_4 = Sequential()
#convolution + max-pooling
model_4.add(Conv2D(4, (4, 4), padding='same', input_shape=(28,28,1)))
model_4.add(Activation('relu'))
model_4.add(MaxPool2D(pool_size=(2, 2)))

model_4.add(Conv2D(8, (4, 4), padding='same'))
model_4.add(Activation('relu'))
model_4.add(MaxPool2D(pool_size=(2, 2)))

model_4.add(Conv2D(16, (4, 4), padding='same'))
model_4.add(Activation('relu'))
model_4.add(MaxPool2D(pool_size=(2, 2)))

#拉平
model_4.add(Flatten())
#隱藏層
model_4.add(Dense(9))
model_4.add(Activation('relu'))
#輸出
model_4.add(Dense(10))
model_4.add(Activation('softmax'))
#組裝
# model_1.compile(loss='mse', optimizer=SGD(lr=0.07), metrics=['accuracy'])
model_4.compile(loss="categorical_crossentropy", optimizer=Adadelta(), metrics=['accuracy'])
#檢視神經網路
model_4.summary()
#訓練
model_4_out = model_4.fit(x_train, y_train, batch_size=100, epochs=10)
#測試資料正確率
score_4 = model_4.evaluate(x_test,y_test)
print('loss :', score_4[0],
      '正確率:', score_4[1])
```
