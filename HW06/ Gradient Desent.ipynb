#  Gradient Desent

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import random
```

### 設定方程式&原始圖形


方程式

$ {L(x) = x^4-3x^3+2x^2+x} $

微分後的方程式

$ {l(x) = 4x^3-9x^2+4x+1} $

```python
def L(x):
    return x**4 - 3*x**3 + 2*x**2 + 1*x
def l(x):
    return 4*x**3 - 9*x**2 + 4*x**1 + 1
```

### 方程式圖形

```python
x = np.linspace(-0.8, 2, 10000)
plt.plot(x,L(x))
```

### 起始值

```python
start = random.uniform(-0.8, 2)
start
```

### 繪製迭代結果

```python
w = start
η = 0.1 

plt.plot(x,L(x))
plt.title("f(x) = x**4 - 3*x**3 + 2*x**2 + 1*x")
loss = l(w)
while True:
    if (loss < 0) and (l(w) >= 0) :
        break
    elif (loss > 0) and (l(w) <= 0) :
        break
    else:
        w = w - η * l(w)
        plt.scatter(w,L(w))
```
